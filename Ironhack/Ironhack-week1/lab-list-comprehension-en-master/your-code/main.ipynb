## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
upper_case = [word.upper() for word in words] # Convert each word to uppercase#
print(upper_case)
```

**Create a new list containing only words with 5 or more letters**

```python
big_words = [word for word in words if len(word) >= 5] # Select words with more than 5 letters#
print(big_words)
```

**Print the first word starting with "t"**

```python
start_t = [word for word in words if word.startswith("t")] # Select words that start with the letter 't'#
print(start_t)
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
squared_numbers = [num**2 for num in range(1,11)] # Create a list of squared numbers from 1 to 10#
print(squared_numbers)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
odd_sqaured = [num**2 for num in range(1,11) if num % 2 != 0]
print(odd_sqaured) # Create a list of squared odd numbers from 1 to 10#
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
multi_8 = [number for number in range(1,1000) if number % 8 == 0] # Create a sugar syntax list of numbers between 1 and 1000 that are multiples of 8#
print(multi_8)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python

print(len(people))  # Calculate how many people are in the room.
```

**How many people have kids**?

```python
kids = [person["n_kids"] for person in people if person["n_kids"] > 0]  # Create a list with the number of kids each person has separately#, in list peopleprint(len(kids)) # sum up the number of peoplem that has kids (has a value more than 0).
print(len(kids)) # sum up the number of people in list of people that has kids (has a value more than 0).
```

**How many kids do they have in total?**

```python
print(sum(kids)) # instead of adding number of people(individual listed items) with kids, we add all the values (ints) together to get the total.
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
more_kids = [person["n_kids"] + 1 if person["name"].endswith("a") else person["n_kids"] for person in people]
for extra, person in enumerate(people):
    
    print(f'{person["name"]} has {more_kids[extra]} kids')
 # Sonia and Lucia have 1 more kid than before.
```
